```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from tqdm.auto import tqdm
import pandas as pd
from bs4 import BeautifulSoup 
from IPython.display import Image
```

```python

def img_source(rlist):
    returnlist=[]
    for i in range(0, len(rlist), 1):
        a= rlist[i].get('src')
        returnlist.append(a)
    return returnlist

def confirm_text (rlist):
    returnlist=[]
    for i in range(0, len(rlist), 1):
        a= rlist[i].text
        aa =a.replace(',',' ')
        returnlist.append(aa)
    return returnlist
```

```python
    url = 'https://whoisnerdy.com/category/%EC%9D%98%EB%A5%98/283/'
    driver = webdriver.Chrome('C:/Users/hi/data01/WebDriver/chromedriver.exe') #브라우저 실행
    driver.get(url)
    time.sleep(3)
```

## 상의추출

```python
top = driver.find_element_by_css_selector("#cha_nav_ul > li:nth-child(6) > a")
top.send_keys(Keys.RETURN)
```

```python
html_doc=BeautifulSoup(html,"html.parser")
```

```python
img=html_doc.select(".thumb_img")
imgs=img_source(img)
imgs
```

```python
productname=html_doc.select(" div > div.description > div.name > a")
productnames=confirm_text(productname)
productnames
```

```python
tops=['상의']*len(imgs)
subcategorys=[]

for i in range(0,len(productnames),1):
    
    if "반팔" in productnames[i]:
        subcategorys.append("반팔")
    elif "가디건" in productnames[i]:
        subcategorys.append("가디건")
    elif "스웨터" in productnames[i]:
        subcategorys.append("스웨터")
    elif "트랙" in productnames[i]:
        subcategorys.append("트레이닝 복")
    elif "스웻셔츠" in productnames[i]:
        subcategorys.append("맨투맨")
    elif "후디" in productnames[i]:
        subcategorys.append("후디")
    elif "긴팔" in productnames[i]:
        subcategorys.append("긴팔")
    elif "드레스" in productnames[i]:
        subcategorys.append("원피스")
    else:
        subcategorys.append("티셔츠")
```
