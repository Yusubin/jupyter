```python
from matplotlib import rc
import matplotlib.font_manager as fm

fm._rebuild()
fm.get_fontconfig_fonts()

# 디렉토리 및 파일 이름에 맞추어 변경
font_location = '../NanumSquareRoundEB.ttf'
font_name = fm.FontProperties(fname=font_location).get_name()
rc('font', family=font_name)
```

```python
f, ax = plt.subplots(1,3,figsize=(16,8))
#explode=[0,0.05]
teen['스트레스 전체'].plot.pie(ax=ax[0], explode=[0.01,0.02], autopct='%1.1f%%')
ax[0].set_title("스트레스를 받은적이 있다.")
teen['자살 전체'].plot.pie(ax=ax[1],  autopct='%1.1f%%')
ax[1].set_title("우울증을 경험한 적이 있다.")
teen['우울 전체'].plot.pie(ax=ax[2],  autopct='%1.1f%%')
ax[2].set_title("자살을 고민한 한 적이 있다.")

plt.show()
```
