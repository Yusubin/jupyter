```python
from surprise import SVD 
#SGD > stochastic 통계적인 Gradient 경사하강법
## 고유값분해(행렬을 더 낮은 차원으로 분해)해서 다시 원본으로 살리면서 잠재적인 값을 추론하는 방법
from surprise import Dataset, Reader ## SVD를 사용하기 위한 데이터셋 만들어주는 클래스
                            ## 사용자, 아이템, 평점
from surprise import accuracy ## RMSE, MAE, cross-validation(k-fold)
from surprise.model_selection import train_test_split ## 훈련/검증 데이터 분류
```

```python
data = Dataset.load_builtin('ml-100k')
data
```

```python
trainset, testset = train_test_split(data, test_size= 0.25, random_state=0)
```

```python
trainset
```

```python
testset #'882' 유저가  '291'이라는 인덱스를 가진 영화에 4.0평점을 주었다. )
```

```python
algo = SVD()
```

```python
algo.fit(trainset)
```

```python
prediction = algo.test(testset)[1]
'user_id:'+prediction.uid+'| iid: '+prediction.iid+'  ===> est:'+ str(prediction.est)
```

```python
## 4. 검증용 데이터로 예측
predictions = algo.test(testset)
predictions ## 하나의 
```

```python
for i in predictions:
    text= 'user_id:'+i.uid+' | iid: '+i.iid+'  ===> est:'+ str(i.est)
    print(text)
```

```python
result2=[ (pred.uid, pred.iid, pred.est) for pred in predictions]
result2
```

```python
uid = str(196)
iid = str(302)
pred = algo.predict(uid, iid)
pred
```

```python
accuracy.rmse(predictions)
```

```python
from surprise.model_selection import cross_validate
```

```python
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True) 
```
