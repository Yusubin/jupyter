```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from tqdm.auto import tqdm
import pandas as pd
from bs4 import BeautifulSoup 
from IPython.display import Image
```

```python

def img_source(rlist):
    returnlist=[]
    for i in range(0, len(rlist), 1):
        a= rlist[i].get('src')
        returnlist.append(a)
    return returnlist

def confirm_text (rlist):
    returnlist=[]
    for i in range(0, len(rlist), 1):
        a= rlist[i].text
        aa =a.replace(',',' ')
        returnlist.append(aa)
    return returnlist
```

```python
    url = 'https://www.musinsa.com/app/cs/faq'
    driver = webdriver.Chrome('../WebDriver/chromedriver.exe') #브라우저 실행
    driver.get(url)
    time.sleep(3)
```

```python
def answer_confirm():
    table= driver.find_element_by_css_selector("body > div.wrap > div.right_area > div.right_contents.faq_contents > table")
    for tr in table.find_elements_by_tag_name("tr"):
        td = tr.find_elements_by_tag_name("td")
    tr=table.find_elements_by_tag_name("tr")
    print (len(tr))
    tr2=len(tr)
    tr2
    product22=[]
    for i in range (1,len(tr),2):
        text="body > div.wrap > div.right_area > div.right_contents.faq_contents > table > tbody > tr:nth-child("+str(i)+")"
        delivery2 = driver.find_element_by_css_selector(text).click()
        html=driver.page_source
        html_doc=BeautifulSoup(html,"html.parser")
        delivery_question2=html_doc.select(" td.faq_a > p")
        product22.append(confirm_text(delivery_question2))

    return product22
```
