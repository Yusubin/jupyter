```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from tqdm.auto import tqdm
```

```python
query = input('검색할 키워드를 입력하세요: ')
url = 'https://shopping.naver.com/home/p/index.naver'
driver = webdriver.Chrome('C:/Users/hi/data01/WebDriver/chromedriver.exe') #브라우저 실행
driver.get(url)
time.sleep(3)
```

```python
#query
search_box = driver.find_element_by_css_selector("#autocompleteWrapper > input.co_srh_input._input.N\=a\:SNB\.search")
search_box.send_keys(query)
search_box.send_keys(Keys.RETURN) #form안에 있는 input의 값을 submit하는 역할!
time.sleep(2)
```

```python
body = driver.find_element_by_tag_name('body')
body.send_keys(Keys.PAGE_DOWN)
```

```python
for i in range(1,10): #숫자를 작게 설정하면, 스크롤 내려가는 것을 볼 수 있음.
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
```

```python
title_list=[]
mask_titles = driver.find_elements_by_css_selector(" div.style_content__2T20F > ul > div > div > li > div.basicList_inner__eY_mq > div.basicList_info_area__17Xyo > div.basicList_title__3P9Q7 > a")
for title in mask_titles:
    print(title.text)
    title_list.append(title.text)
title_list
```

```python
rink_list=[]
rink_titles = driver.find_elements_by_css_selector(" div.basicList_title__3P9Q7 > a")
for r in rink_titles:
    print(r.get_attribute('href'))
    rink_list.append(r.get_attribute('href'))
rink_list
```

```python
price= driver.find_elements_by_css_selector("strong > span > span.price_num__2WUXn")
price_list=[]
for p in price:
    print(p.text)
    price_list.append(p.text)
price_list
```

```python
review=driver.find_elements_by_css_selector(" li > div.basicList_inner__eY_mq > div.basicList_info_area__17Xyo > div.basicList_etc_box__1Jzg6 > a > em")
review_list=[]
for r in review:
    print(r.text)
    review_list.append(r.text)
review_list
```

```python
tool=driver.find_elements_by_css_selector(" ul > div > div > li > div.basicList_inner__eY_mq > div.basicList_info_area__17Xyo > div.basicList_desc__2-tko > div.basicList_detail_box__3ta3h > a:nth-child(3)")
tool_list=[]
for t in tool:
    print(t.text)
    tool_list.append(t.text)
tool_list
```

```python
thumbnails=driver.find_elements_by_css_selector("#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.style_content_wrap__1PzEo > div.style_content__2T20F > ul > div > div > li > div > div.basicList_img_area__a3NRA > div > a > img")
thumbnail = []
for img in thumbnails :
    print(img.get_attribute('src'))
    thumbnail.append(img.get_attribute('src'))
thumbnail
```

```python
import pandas as pd
```

```python
tool_list=tool_list[:20]
title_list=title_list[:20]
review_list=review_list[:20]
thumbnail=thumbnail[:20]
price_list=price_list[:20]
rink_list=rink_list[:20]
```

```python
naver_search_df=pd.DataFrame({'title':title_list, 'img':thumbnail, 'price':price_list, 'tool': tool_list,'review':review_list,'rink':rink_list})
naver_search_df
```

```python
# 이미지 저장할 폴더 생성

import os

# path_folder의 경로는 각자 저장할 폴더의 경로를 적어줄 것(ex.img_download)
path_folder = './'

if not os.path.isdir(path_folder):
    os.mkdir(path_folder)


# 이미지 다운로드

from urllib.request import urlretrieve

i = 0

for link in thumbnail:          
    i += 1
    urlretrieve(link, path_folder + f'{i}.jpg')   
```
