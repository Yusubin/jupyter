```python
import requests
import pandas as pd
import numpy as np
import pickle
def connect(url,servicekey, perpage):
    headers = {
        'accept': 'application/json',
        'Authorization': servicekey,
    }

    params = (
        ('page', '1'),
        ('perPage', perpage),
        ('serviceKey',servicekey),
    )
    response = requests.get(url, headers=headers, params=params)
    return response.json()
```

```python
json_data= connect(url='https://api.odcloud.kr/api/15066516/v1/uddi:507e01f5-76ec-42ff-96a5-8b6ff9ce554e', servicekey= '-',perpage=300)
json_data
```

```python
na=json_data['data']
```

```python
def genre_sel(i):
    genre=' '
    g = na[i].get('종류(01한식,02중식,03일식,04양식,05기타외국음식,06디저트/카페)')
    if g==1:
        genre='한식'
    elif g==2:
        genre='중식'
    elif g==3:
        genre='일식'
    elif g==4:
        genre='양식'
    elif g==5:
        genre='기타외국음식'
    else:
        genre='디저트/카페'

    return genre
```

```python
list1=[]
for no in range(len(na)):
    name = na[no].get('업소명')
    location = na[0].get('주소1')+" "+na[0].get('주소2')
    tel =  na[no].get('전화번호')
    theme =  na[no].get('테마(국가)')
    etc =  na[no].get('사장님이자랑하는내가게한마디')
    genre = genre_sel(no)
    
    one_row=[]
    one_row.append(name)
    one_row.append(genre)
    one_row.append(theme)
    one_row.append(location)
    one_row.append(tel)
    one_row.append(etc)
    list1.append(one_row)
list1
```

```python

with open('best-youngsan.json', 'wb') as fp:
    pickle.dump(list1, fp)
```
