```python
#판단선(회귀선): 예측과 분류의 기준이 되는 선!
#판단선은 방정식, 1차방정식이라고한다면 기울기와 절편이 필요하다
```

```python
# 위의 그래프는 무엇을 나타내는가 서울의 인구수(x축)에 따른 cctv구별 설치수를 나타낸다. 
# cctv를 어느구에 설치해야하는가 ? 
# 도봉구, 서대문구 등등 회귀선보다 아래에 있는 지역들에 cctv를 더 설치해야한다. ==>오차가 큰순으로~~.... 
# 판단선 위에 위치한, (ex) 강남구 ... 는 이미 cctv설치가 많이 되어있어 더욱 설치 안해도 될것같다. 
# 판단 기준... : 인구수와 cctv가 관련이 있더라... 
```

```python
#판단선들의 값들이 필요 
#현재 값들이 필요
#저 둘간의 오차를 구하기.... 
data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))
```

```python
df_sort=data_result.sort_values(by='오차', ascending=False)
df_sort
```

```python
data_result_crime=pd.merge(df, data_result, on='구별')
data_result_crime.info()
```

```python
data_result_crime['절도 발생']=data_result_crime['절도 발생'].str.replace(',','')
data_result_crime['절도 검거']=data_result_crime['절도 검거'].str.replace(',','')
data_result_crime['폭력 발생']=data_result_crime['폭력 발생'].str.replace(',','')
data_result_crime['폭력 검거']=data_result_crime['폭력 검거'].str.replace(',','')

data_result_crime['절도 발생'] = pd.to_numeric(data_result_crime['절도 발생'])
data_result_crime['절도 검거'] = pd.to_numeric(data_result_crime['절도 검거'])
data_result_crime['폭력 발생'] = pd.to_numeric(data_result_crime['폭력 발생'])
data_result_crime['폭력 검거'] = pd.to_numeric(data_result_crime['폭력 검거'])
```

```python
data_result_crime
```

```python
data_result_crime['살인 미검거'] = data_result_crime['살인 발생']-data_result_crime['살인 검거']
data_result_crime['강도 미검거'] = data_result_crime['강도 발생']-data_result_crime['강도 검거']
data_result_crime['강간 미검거'] = data_result_crime['강간 발생']-data_result_crime['강간 검거']
data_result_crime['절도 미검거'] = data_result_crime['절도 발생']-data_result_crime['절도 검거']
```

```python
data_result_crime.drop(columns=['강도 검거', '절도 검거', '폭력 검거', '강간 검거','살인 검거' ], inplace=True)
```
