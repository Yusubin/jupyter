```python
# Python3 샘플 코드 #


import requests

url = 'http://apis.data.go.kr/6430000/goodRestaService/getGoodResta'
params ={'serviceKey' : '-', 'currentPage' : '1', 'perPage' : '1000' }

response = requests.get(url, params=params)
```

```python
json_data = response.json()
json_data
```

```python
na=json_data['body']
na[0].get('SIGUN_NM')#지역이름
```

```python
location = na[0].get('SIGUN_NM')+" "+na[0].get('DETAIL_ADRES')
location
```

```python
list1=[]
for no in range(len(na)):
    name = na[no].get('BSSH_NM')
    location = na[0].get('SIGUN_NM')+' '+na[0].get('DETAIL_ADRES')
    tel =  na[no].get('TELNO')
    etc =  na[no].get('RM')
    
    one_row=[]
    one_row.append(name)
    one_row.append(location)
    one_row.append(tel)

    one_row.append(etc)
    list1.append(one_row)
list1
```

```python
import pickle
with open('best-bob.json', 'wb') as fp:
    pickle.dump(list1, fp)
```

```python
with open('best-bob.json', 'rb') as fp:
    json_sub_data=pickle.load(fp)
```
