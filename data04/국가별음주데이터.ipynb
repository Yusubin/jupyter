```python
#컬럼 == 변수 == 특성(feature, 피쳐) 
#속성(property, attribute) == entry
#열 == field

#저장할데이터의 컬럼을 선정하는 것이 중요.... 
#칼럼을 재가공
##피쳐 엔지니어링

#컬럼간의 상관정도를 수치로 표현하여 칼럼간의 관계성을 고려해볼수있음... 
#상관도: 상관이 있는정도 (Correlation)
#상관계수:상관정도가 수치로 표현된 것(corr)

## 상관 =/= 연관

## 상관:관계가 있긴 있음을 나타내는 정도(인과관계가 없음)
## 연관: 관계가 있고, 하나가 다른 하나에 영향을 미치는 정도(인과관계가 있음.)

#상관도를 구해볼때 사용하는 방법
#1) corr함수를 이용
#2) 그래프를 이용~ heatmap(), pairplot()
```

```python
df[['wine_servings','beer_servings']].corr() #상관도를 구하고 수치보는법
# 0.7 ~ 1 : 높은 상관관계   
# 0.5 ~ 0.7 : 중간정도의 상관관계
# 0.3 ~ 0.5 : 낮은 상관관계를 갖는다. 
# 0 ~ 0.3 : 상관이 거의 없다
# +: 양의 상관관계      ....  -:음의 상관관계
```

```python
c = df.corr()#피어슨 상관계수
```

```python
# [살균제 양, 살균제 사용연령, 기후, 방의 크기, 살균제 사용기간,층의 높이, 사망여부]
# 기후, 층의 높이, 방의크기 는 상관이 없는 항목처럼 보임
# 기후, 층, 방 크기...은  corr을 구해보면 아주낮은 상관도를 보임 0.0001.
```

```python
import seaborn as sns

# corr 행렬 히트맵을 시각화합니다.
cols = ['beer', 'spirit', 'wine', 'alcohol'] # 그래프 출력을 위한 cols 이름을 축약합니다.
sns.set(font_scale=1.0)
```

```python
hm = sns.heatmap(c.values,
            cbar=True,
            annot=True, 
            square=True,
            fmt='.2f',
            annot_kws={'size': 10},
            yticklabels=cols,
            xticklabels=cols)

plt.tight_layout()
plt.show()
```
