```python
import requests
import time
from urllib.parse import quote
import pandas as pd
import urllib.request


client_id = '-'
client_secret = '-'
headers = {'X-Naver-Client-Id':client_id, 'X-Naver-Client-Secret':client_secret}

url_base="https://openapi.naver.com/v1/search/blog.json?query="
keyword = quote(input("검색 키워드를 입력하세요. : "))
keyword_number=input("몇 번째부터 검색할까요? : ")
display=input("몇 개까지 검색할까요? : ")


parameters = "?query = %s&start=%s&display=%s" %(keyword, keyword_number, display)
url = url_base + parameters
print(url)
result = requests.get(url,headers = headers).json()                  #결과값을 변수로 지정
result 
```

```python
dic=result['items']
dic
```

```python
dic[0].get('title')
```

```python
list1=[]
for no in range(0,len(dic),1):
    title = dic[no].get('title')
    name = dic[no].get('bloggername')
    location = dic[no].get('bloggerlink')
    postdate =  dic[no].get('postdate')
    description =  dic[no].get('description')
    
    one_row=[]
    one_row.append(title)
    one_row.append(name)
    one_row.append(location)
    one_row.append(postdate)
    one_row.append(description)
    list1.append(one_row)
list1
```

```python
import pickle
with open('best-youngsan-naver.json', 'wb') as fp:
    pickle.dump(list1, fp)
```

```python
df1=pd.read_csv('youngsan-blog.csv', encoding='utf8')
df1.drop(columns=['Unnamed: 0'], inplace=True)
df1
```

```python
df1['postdate']= df1['postdate'].astype('str')
df1['date']=pd.to_datetime(df1['postdate'])
```

```python
count_y=[0]*3
year_list= df1['date'].dt.year
month_list= df1['date'].dt.month
count_m=[0]*12
for i in year_list:
    if i==2022:
        count_y[0] +=1
    elif i==2021:
        count_y[1] +=1
    elif i<=2020:
        count_y[2] +=1

        
for i in(month_list):
    for j in range(1,13,1):
        if(i==j):
            count_m[j-1]+=1
        
count_y

# count_m
```
