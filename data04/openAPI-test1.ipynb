```python
import requests

headers = {
    'accept': 'application/json',
    'Authorization': '-',
}

params = (
    ('page', '1'),
    ('perPage', '10'),
    ('serviceKey', '-'),
)

## decode key가 적용되었음.
## api마다 연결이 되지 않는다고 한다면, encode, decode를 바꾸어 테스트해볼 것.
url='https://api.odcloud.kr/api/15069309/v1/uddi:973ad0df-617a-4565-8e82-8ea7869e75d4'
response = requests.get(url, headers=headers, params=params)
response
```

```python
json_data = response.json()
json_data
```

```python
import pickle
with open('covid-go.kr.json', 'wb') as fp:
    pickle.dump(json_data, fp)
    #fp=file.open('covid-go.kr.json', 'wb') 
    #fp.close
```

```python
list1=[]
for no in range(10):
    com = data_values[no].get('언론사')
    content = data_values[no].get('본문')
    title = data_values[no].get('제목')
    addr = data_values[no].get('원본주소')
    date = data_values[no].get('일자')
    field = data_values[no].get('통합 분류1')
    one_row=[]
    one_row.append(field)
    one_row.append(title)
    one_row.append(com)

    one_row.append(content)
    one_row.append(date)
    one_row.append(addr)
    list1.append(one_row)
    
```
