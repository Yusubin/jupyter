```python
import warnings
warnings.filterwarnings(action='ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from tqdm.auto import tqdm
import pandas as pd
from bs4 import BeautifulSoup 
from IPython.display import Image
```

```python
def connect():
    driver =  webdriver.Chrome('C:/Users/hi/data01/WebDriver/chromedriver.exe')
    return driver
```

```python
def coffee_bean(no):
    #driver2 = None # => 변수 새로 할당하는 경우!!
    driver2.get("https://www.coffeebeankorea.com/store/store.asp")
    time.sleep(1)
    func = "storePop2('" + str(no) + "')"
    shop = []
    try:
        # 스크립트 실행
        driver2.execute_script(func)
        time.sleep(1)
        
        # 실행한 결과 html 가지고 오세요.
        html2 = driver2.page_source
        
        # bsparser설정
        doc2 = BeautifulSoup(html2, 'html.parser')
        
        # 타이틀 검색
        title3 = doc2.select('div.store_txt > h2')
        title4 = title3[0].string
        print('title4>> ', title4)
        shop.append(title4)
        
        # 주소 검색
        info2 = doc2.select('div.store_txt > table.store_table > tbody > tr > td')
        time2 = info2[0].getText()
        print('time2>> ', time2)
        shop.append(time2)
    
        parking2 = info2[1].getText()
        print('parking2>> ', parking2)
        shop.append(parking2)
        addr2 = info2[2].getText()
        print('addr2>> ', addr2)
        shop.append(addr2)
        
        # 전화번호 검색
        tel2 = info2[3].getText()
        print("tel2",tel2)
        shop.append(tel2)
    except:
        print('해당번호없음')
        shop.append('null')

    return shop
```

```python
result2 = []
wd = None
wd = connect()
for no in range(1, 200):
    print('=========' + str(no) + '===========')
    shop = coffee_bean(no)
    print(shop)
    if shop != '':
        result2.append(shop)
        print('length>> ', len(result2))
```
