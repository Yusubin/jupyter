```python
import seaborn as sns
import pandas as pd
import numpy as np #
df = sns.load_dataset("titanic")
df.head()
```

```python
# 나이가 30살 이상 남자 승객 조건 필터링and 'sex'='male']
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
df[(df['age']>=30)& (df['sex']=='male')].sort_values(by="fare", ascending=False).head(10)
```

```python
# 나이가 20살 이상 40살 미만인 승객
# pclass가 1등급 혹은 2등급인 승객
# 열(column)은 survived, pclass, age, fare 만 나오게 출력
# 10개만 출력
agetwenty_forty=df[(df['age']>=20)& (df['age']<40)].head(10)
pclass_one_two=df[df['pclass'].isin([1, 2])].head(10)
print("# 나이가 20살 이상 40살 미만인 승객")
print(agetwenty_forty[['survived', 'pclass', 'age', 'fare']])
print("--------------------------------------------------------")
print("# pclass가 1등급 혹은 2등급인 승객")
print(pclass_one_two[['survived', 'pclass', 'age', 'fare']])
```
