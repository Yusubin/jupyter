```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([1,2,3,4,5,6])
ypoints=np.array([30,50,70,20,20,70])

plt.plot(xpoints,ypoints, marker='*', linestyle='dotted')#직선그래프
plt.ylabel('count')
plt.xlabel('numbers of students')
# plt.legend()
plt.show()
```

```python
xpoints=np.array([1,2,3,4,5,6])
ypoints=np.array([30,50,70,20,20,70])

plt.scatter(xpoints,ypoints)#산점도> 점으로 그려줌

plt.show()
```

```python
import random
x=[]
y=[]

for i in range(1,102,1):
    x.append(i)
    y.append(random.randint(1,100))

x
y
xpoints=np.array(x)
ypoints=np.array(y)
plt.scatter(xpoints,ypoints)#산점도> 점으로 그려줌
plt.show()

# plt.plot(x,y)#직선그래프
# plt.show()
```
