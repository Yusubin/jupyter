```python
#5.df 맨 마지막에 price 항목을 추가(값은 임의대로 넣음 , int)
import random
random_price=[]
for i in range (df.shape[1]):
    random_price.append(random.randint(10000,70000))
df['price']=random_price
df
```

```python
#6. df의 맨 마지막에 percent항목을 추가 ( "" )
random_per=[]
for i in range (df.shape[1]):
    random_per.append(random.uniform(0.0,1.0))

df['percent']=random_per
df
```

```python
#7.df의 맨 마지막에 sale항목을 추가 (price*percent)
df['sale']=df['percent']*df['price']
df
```
