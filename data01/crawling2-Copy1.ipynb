```python
from urllib import request
```

```python
from bs4 import BeautifulSoup
```

```python
con =request.urlopen('https://finance.naver.com/sise/sise_quant.naver')
con 
#http response ( header + body)
#header: http result status code (ok:200, not found:404  ,error:500,  permission:403)
```

```python
con.status #status code
```

```python
doc=BeautifulSoup(con,'html.parser')
doc

```

```python
title_list=doc.select('#boxEtf > div > div > table > tbody > tr.first > td:nth-child(5) > span')
tit_list=[]
for i in range(0, len(title_list), 1):
    tit_list.append(title_list[i].text)

tit_list
```

```python
siger_list=doc.select('#content > div.lower_content>div.pd_unit.pd_unit04 > div>ul>li>a> div>p.brand')
brand_list=[]
for i in range(0, len(siger_list), 1):
    brand_list.append(siger_list[i].text)
brand_list
```

```python
sale_list=doc.select('#content > div.lower_content>div.pd_unit.pd_unit04 > div>ul>li>a> div> p.price.font_t > span.percent')
percent_list=[]
for i in range(0, len(sale_list), 1):
    percent_list.append(sale_list[i].text)

percent_list
```

```python
p_list=doc.select('#content > div.lower_content>div.pd_unit.pd_unit04 > div>ul>li>a> div> p.price.font_t > span:nth-child(2)')
price_list=[]
for i in range(0, len(p_list), 1):
    aa =int(p_list[i].text.replace(',',''))
    price_list.append(aa)

price_list
```

```python

album_file =open('album.csv','w')
album_list=[]
album_file.write("title,brand,price,percent"+"\n")

for i in range(0,len (price_list),1):
    data=tit_list[i]+","+brand_list[i]+","+str(price_list[i])+","+percent_list[i]+"\n"
    album_file.write(data)
    
album_file.close()  
```
